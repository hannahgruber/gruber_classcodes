```python
# importing code/packages

import numpy as np
import matplotlib.pyplot as plt
from mesa_web import read_profile
```

```python
# reading profile and obtaining values to graph

final_profile=read_profile('/Users/hannahgruber/Documents/400A/MESA/MESA_sun_analog/profile8.data')
mass=final_profile['mass']
radius=final_profile['radius']
M=final_profile['star_mass'] # single global value
R=final_profile['photosphere_r'] # single global value
P=final_profile['pressure']
T=final_profile['logT']
XH=final_profile['h1']
Yalpha=final_profile['he4']
ZC=final_profile['c12']
outer_L=final_profile['photosphere_L']
```

```python
# plotting mass versus radius

fig, ax = plt.subplots(figsize=(5,5))
plt.xlabel("Radius [Rsun]")
plt.ylabel("Mass [Msun]")
plt.title("m(r) plot of Sun analog")
plt.plot(radius,mass)
plt.show()

''' 
Plotting mass versus radius feels like a natural choice that will show where
the most mass is concentrated within the sun. The greatest change in mass happens
between approximately r = 0.1 - 0.5 Rsun, meaning that most of the mass lies there.
'''
```

```python
# plotting mass versus pressure

fig, ax = plt.subplots(figsize=(5,5))
plt.xlabel("Pressure [dyn/cm^2]")
plt.ylabel("Mass [Msun]")
plt.title("m(P) plot of Sun analog")
plt.plot(P,mass)
plt.show()

'''
I chose to plot pressure because of what an important factor it is in stellar structure.
My plot shows that most of the mass is concentrated at lower pressure. This surprised
me, as I would expext more mass concentrated in the center, with the highest pressure. 
'''
```

```python
# plotting mass versus temp

fig, ax = plt.subplots(figsize=(5,5))
plt.xlabel("Temperature [K]")
plt.ylabel("Mass [Msun]")
plt.title("m(T) plot of Sun analog")
plt.plot([10**t for t in T],mass)
plt.show()

''' 
I chose to plot temperature due to how the temperature increases at smaller radius. In
accordance with that, this is almost an inverse of my m(r) plot, with some changes to the
shape of the plot and its slope. I was provided with log10(temperature) from the MESA model,
but I believe it is more valuable to plot temperature itself (and that plot ended up mostly
as a plateau, so not very interesting anyway).
'''
```

```python
# plotting mass versus hydrogen mass fraction

fig, ax = plt.subplots(figsize=(5,5))
plt.xlabel("1H mass fraction")
plt.ylabel("Mass [Msun]")
plt.title("m(X) plot of Sun analog")
plt.plot(XH,mass)
plt.show()

''' 
I chose to plot different elemental mass fractions, the first of which being Hydrogen,
to see a distribution of these elements. While it is an interesting variable due to its
nature as a mass measurement itself, I think it is still an interesting metric. We can 
see from this plot that most of the star's mass has a high (>.6) mass fraction of Hydrogen.
'''
```

```python
# plotting mass versus helium-4 mass fraction

fig, ax = plt.subplots(figsize=(5,5))
plt.xlabel("4He mass fraction")
plt.ylabel("Mass [Msun]")
plt.title("m(Y) plot of Sun analog")
plt.plot(Yalpha,mass)
plt.show()

''' 
Helium is the second element whose mass fraction I am plotting. From this plot you can see
that most of the mass has a low (>.4) mass fraction. Interestingly, this matches quite 
nicely with the behavior of the Hydrogen mass fraction, with their shares of the mass
fractions adding up to almost one.
'''
```

```python
# plotting mass versus carbon mass fraction

fig, ax = plt.subplots(figsize=(5,5))
plt.xlabel("12C mass fraction")
plt.ylabel("Mass [Msun]")
plt.title("m(Z_C) plot of Sun analog")
plt.plot(ZC,mass)
plt.show()

''' 
I chose the Carbon mass fraction to represent the general metallicity of the star. While not
as interesting, this graph shows that the metallicity is consistently low throughout the star.
'''
```

```python
# pressure versus radius

fig, ax = plt.subplots(figsize=(5,5))
plt.xlabel("Radius [Rsun]")
plt.ylabel("Pressure [dyn/cm^2]")
plt.title("P(r) plot of Sun analog")
plt.plot(radius,P)
plt.show()

''' 
This shows that pressure is highest in the center of the star. The central pressure is
approximately 2.27e17 dyn/cm^2. This is a couple of orders of magnitude higher than Onno Pols' 
estimate of 7e15 dyn/cm^2.
'''
```
